```python
def get_data_between_substring(text, substring):
    data_between = []
    start_index = 0
    while True:
        end_index = text.find(substring, start_index)
        if end_index == -1:
              break
        data_between.append(text[start_index:end_index])
        start_index = end_index + len(substring)
    return data_between

# Example usage
# text = "This is some text with a substring in it. Here is another substring.This2 is some text with a substring in it. Here2 is another substring."
text = index
substring = "_3-95 _2pim _a6-h _a6-i"
data_between = get_data_between_substring(text, substring)
print(data_between)
```
